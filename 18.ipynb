```python
# An√°lisis ARIMA para el logaritmo del IPD (Ingreso Personal Disponible)
# Versi√≥n con carga manual de archivo

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from google.colab import files
import io
import warnings
warnings.filterwarnings('ignore')

# Configurar el estilo de los gr√°ficos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# PASO 1: CARGAR ARCHIVO DE DATOS
print("=" * 60)
print("PASO 1: CARGA DE DATOS")
print("=" * 60)

print("\nüìÅ Por favor, sube el archivo con los datos macroecon√≥micos.")
print("El archivo debe contener las columnas: YEAR, GDP, PDI, PCE, PROFITS, DIVIDENDS")
print("\nFormatos aceptados: .txt, .csv, .xlsx\n")

# Subir archivo
uploaded = files.upload()

# Obtener el nombre del archivo subido
filename = list(uploaded.keys())[0]
print(f"\n‚úÖ Archivo '{filename}' cargado exitosamente!")

# Leer el archivo seg√∫n su extensi√≥n
try:
    if filename.endswith('.csv'):
        # Para archivos CSV
        df_raw = pd.read_csv(io.StringIO(uploaded[filename].decode('utf-8')))
    elif filename.endswith('.txt'):
        # Para archivos TXT (asumiendo que est√° delimitado por espacios o tabs)
        # Primero intentamos leer las primeras l√≠neas para detectar el formato
        content = uploaded[filename].decode('utf-8')
        lines = content.strip().split('\n')

        # Buscar la l√≠nea donde empiezan los datos (despu√©s de los headers)
        data_start = 0
        for i, line in enumerate(lines):
            if line.strip() and line.strip()[0].isdigit():
                data_start = i
                break

        # Leer los datos
        data_lines = []
        for line in lines[data_start:]:
            if line.strip() and line.strip()[0].isdigit():
                data_lines.append(line.strip().split())

        # Crear DataFrame
        columns = ['YEAR', 'GDP', 'PDI', 'PCE', 'PROFITS', 'DIVIDENDS']
        df_raw = pd.DataFrame(data_lines, columns=columns)

        # Convertir columnas num√©ricas
        for col in columns[1:]:
            df_raw[col] = pd.to_numeric(df_raw[col], errors='coerce')

    elif filename.endswith('.xlsx') or filename.endswith('.xls'):
        # Para archivos Excel
        df_raw = pd.read_excel(io.BytesIO(uploaded[filename]))
    else:
        raise ValueError("Formato de archivo no soportado")

    print("\nüìä Datos cargados correctamente!")
    print(f"Dimensiones: {df_raw.shape[0]} filas √ó {df_raw.shape[1]} columnas")
    print("\nPrimeras 5 filas del archivo:")
    print(df_raw.head())

except Exception as e:
    print(f"\n‚ùå Error al leer el archivo: {e}")
    print("\nPor favor, aseg√∫rate de que el archivo tenga el formato correcto.")
    raise

# Verificar que existe la columna PDI
if 'PDI' not in df_raw.columns:
    print("\n‚ùå Error: No se encontr√≥ la columna 'PDI' en el archivo.")
    print(f"Columnas encontradas: {list(df_raw.columns)}")
    raise ValueError("La columna PDI es necesaria para el an√°lisis")

# Procesar los datos
print("\n" + "=" * 60)
print("PROCESAMIENTO DE DATOS")
print("=" * 60)

# Crear DataFrame con el formato necesario
df = pd.DataFrame()

# Manejar la columna de fechas
if 'YEAR' in df_raw.columns:
    df['Year_Quarter'] = df_raw['YEAR'].astype(str)
else:
    print("‚ö†Ô∏è No se encontr√≥ columna 'YEAR'. Creando √≠ndice temporal...")
    df['Year_Quarter'] = range(len(df_raw))

# Copiar PDI
df['PDI'] = df_raw['PDI'].astype(float)

# Convertir a fechas trimestrales si es posible
try:
    # Intentar convertir formato a√±o.trimestre a fecha
    df['Date'] = pd.to_datetime(df['Year_Quarter'].str.replace('.', 'Q'), format='%YQ%q')
    df.set_index('Date', inplace=True)
    print("‚úÖ Fechas convertidas a formato trimestral")
except:
    print("‚ö†Ô∏è No se pudo convertir a fechas trimestrales. Usando √≠ndice num√©rico.")
    df.index = range(len(df))

# Eliminar valores faltantes
df = df.dropna(subset=['PDI'])
print(f"\nüìà Datos procesados: {len(df)} observaciones v√°lidas")

# Calcular el logaritmo del IPD
df['log_PDI'] = np.log(df['PDI'])

print("\nEstad√≠sticas del IPD:")
print(df['PDI'].describe())

# PASO 2: AN√ÅLISIS EXPLORATORIO Y VISUALIZACI√ìN
print("\n" + "=" * 60)
print("PASO 2: AN√ÅLISIS EXPLORATORIO")
print("=" * 60)

# Visualizar la serie original y su logaritmo
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# IPD original
ax1.plot(df.index, df['PDI'], linewidth=2, color='blue', marker='o', markersize=3, alpha=0.7)
ax1.set_title('IPD (Ingreso Personal Disponible) - Serie Original', fontsize=14)
ax1.set_ylabel('Miles de millones de $ (1987)', fontsize=12)
ax1.grid(True, alpha=0.3)

# Logaritmo del IPD
ax2.plot(df.index, df['log_PDI'], linewidth=2, color='green', marker='o', markersize=3, alpha=0.7)
ax2.set_title('Logaritmo del IPD', fontsize=14)
ax2.set_ylabel('log(IPD)', fontsize=12)
ax2.set_xlabel('Per√≠odo', fontsize=12)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Estad√≠sticas descriptivas
print("\nEstad√≠sticas descriptivas del log(IPD):")
print(df['log_PDI'].describe())

# PASO 3: PRUEBAS DE ESTACIONARIEDAD
print("\n" + "=" * 60)
print("PASO 3: PRUEBAS DE ESTACIONARIEDAD")
print("=" * 60)

def test_stationarity(series, series_name):
    """Realizar pruebas ADF y KPSS para estacionariedad"""
    print(f"\nPruebas para: {series_name}")
    print("-" * 40)

    # Prueba ADF (Dickey-Fuller Aumentada)
    adf_result = adfuller(series, autolag='AIC')
    print("\nüîç Prueba ADF (H0: Serie tiene ra√≠z unitaria - No estacionaria)")
    print(f"  Estad√≠stico ADF: {adf_result[0]:.4f}")
    print(f"  p-valor: {adf_result[1]:.4f}")
    print(f"  Valores cr√≠ticos:")
    for key, value in adf_result[4].items():
        print(f"    {key}: {value:.4f}")

    if adf_result[1] < 0.05:
        print("  ‚úÖ Conclusi√≥n: Rechazamos H0 - La serie ES estacionaria")
    else:
        print("  ‚ùå Conclusi√≥n: No rechazamos H0 - La serie NO es estacionaria")

    # Prueba KPSS
    kpss_result = kpss(series, regression='ct', nlags='auto')
    print("\nüîç Prueba KPSS (H0: Serie es estacionaria)")
    print(f"  Estad√≠stico KPSS: {kpss_result[0]:.4f}")
    print(f"  p-valor: {kpss_result[1]:.4f}")
    print(f"  Valores cr√≠ticos:")
    for key, value in kpss_result[3].items():
        print(f"    {key}: {value:.4f}")

    if kpss_result[1] > 0.05:
        print("  ‚úÖ Conclusi√≥n: No rechazamos H0 - La serie ES estacionaria")
    else:
        print("  ‚ùå Conclusi√≥n: Rechazamos H0 - La serie NO es estacionaria")

    return adf_result[1], kpss_result[1]

# Probar estacionariedad de log(IPD)
adf_pval, kpss_pval = test_stationarity(df['log_PDI'], 'log(IPD) - Nivel')

# PASO 4: DIFERENCIACI√ìN SI ES NECESARIA
print("\n" + "=" * 60)
print("PASO 4: DIFERENCIACI√ìN")
print("=" * 60)

# Determinar si es necesaria la diferenciaci√≥n
if adf_pval > 0.05 or kpss_pval < 0.05:
    print("‚ö†Ô∏è La serie en nivel no es estacionaria. Aplicando primera diferencia...")
    d = 1
else:
    print("‚úÖ La serie en nivel es estacionaria. No es necesaria la diferenciaci√≥n.")
    d = 0

# Primera diferencia
df['d_log_PDI'] = df['log_PDI'].diff()

# Visualizar serie diferenciada
plt.figure(figsize=(12, 5))
plt.plot(df.index[1:], df['d_log_PDI'][1:], linewidth=2, color='red', marker='o', markersize=3, alpha=0.7)
plt.title('Primera Diferencia del log(IPD)', fontsize=14)
plt.ylabel('Œîlog(IPD)', fontsize=12)
plt.xlabel('Per√≠odo', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
plt.show()

# Probar estacionariedad de la serie diferenciada
if d == 1:
    adf_pval_d, kpss_pval_d = test_stationarity(df['d_log_PDI'].dropna(),
                                                 'log(IPD) - Primera Diferencia')

# PASO 5: IDENTIFICACI√ìN DEL MODELO (ACF y PACF)
print("\n" + "=" * 60)
print("PASO 5: IDENTIFICACI√ìN DEL MODELO ARIMA")
print("=" * 60)

# Determinar qu√© serie usar para ACF/PACF
if d == 1:
    series_for_acf = df['d_log_PDI'].dropna()
    series_name = "Primera Diferencia"
else:
    series_for_acf = df['log_PDI']
    series_name = "Serie en Nivel"

# Gr√°ficos ACF y PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# ACF
plot_acf(series_for_acf, lags=min(20, len(series_for_acf)//4), ax=ax1, alpha=0.05)
ax1.set_title(f'Funci√≥n de Autocorrelaci√≥n (ACF) - {series_name}', fontsize=14)
ax1.set_xlabel('Rezagos')

# PACF
plot_pacf(series_for_acf, lags=min(20, len(series_for_acf)//4), ax=ax2, alpha=0.05)
ax2.set_title(f'Funci√≥n de Autocorrelaci√≥n Parcial (PACF) - {series_name}', fontsize=14)
ax2.set_xlabel('Rezagos')

plt.tight_layout()
plt.show()

print("\nüìä Interpretaci√≥n de ACF y PACF:")
print("- Si ACF decae gradualmente y PACF se corta ‚Üí Modelo AR")
print("- Si PACF decae gradualmente y ACF se corta ‚Üí Modelo MA")
print("- Si ambos decaen gradualmente ‚Üí Modelo ARMA")

# PASO 6: SELECCI√ìN Y ESTIMACI√ìN DE MODELOS
print("\n" + "=" * 60)
print("PASO 6: ESTIMACI√ìN DE MODELOS ARIMA")
print("=" * 60)

# Funci√≥n para ajustar y evaluar modelos ARIMA
def fit_arima_model(data, order, seasonal_order=None):
    """Ajustar modelo ARIMA y mostrar resultados"""
    try:
        if seasonal_order:
            model = ARIMA(data, order=order, seasonal_order=seasonal_order)
        else:
            model = ARIMA(data, order=order)

        fitted_model = model.fit()
        print(f"\nüìà Modelo ARIMA{order}:")
        print(f"   AIC: {fitted_model.aic:.2f}")
        print(f"   BIC: {fitted_model.bic:.2f}")
        print(f"   Log-likelihood: {fitted_model.llf:.2f}")

        return fitted_model
    except Exception as e:
        print(f"\n‚ùå Error al ajustar ARIMA{order}: {str(e)[:50]}...")
        return None

# Determinar el valor de d basado en las pruebas
if d == 0:
    d_values = [0]
else:
    d_values = [1]

# Probar diferentes especificaciones ARIMA
models = {}
model_specs = []

# Generar especificaciones de modelos
for d_val in d_values:
    for p in range(0, 3):
        for q in range(0, 3):
            if p == 0 and q == 0:
                continue
            model_specs.append((p, d_val, q))

print(f"üîç Probando {len(model_specs)} especificaciones de modelos ARIMA...")

for spec in model_specs:
    model = fit_arima_model(df['log_PDI'], spec)
    if model:
        models[spec] = model

# Seleccionar el mejor modelo basado en AIC
if models:
    best_model_spec = min(models.keys(), key=lambda x: models[x].aic)
    best_model = models[best_model_spec]

    print(f"\n{'='*60}")
    print(f"üèÜ MEJOR MODELO: ARIMA{best_model_spec}")
    print(f"{'='*60}")
    print(best_model.summary())
else:
    print("\n‚ùå No se pudo ajustar ning√∫n modelo. Revisa los datos.")
    raise ValueError("No se pudo ajustar ning√∫n modelo ARIMA")

# PASO 7: DIAGN√ìSTICO DEL MODELO
print("\n" + "=" * 60)
print("PASO 7: DIAGN√ìSTICO DEL MODELO")
print("=" * 60)

# Residuos
residuals = best_model.resid

# Gr√°ficos de diagn√≥stico
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Residuos vs tiempo
axes[0, 0].plot(residuals, marker='o', markersize=4, alpha=0.7)
axes[0, 0].set_title('Residuos vs Tiempo')
axes[0, 0].axhline(y=0, color='red', linestyle='--')
axes[0, 0].set_xlabel('Observaci√≥n')
axes[0, 0].set_ylabel('Residuos')
axes[0, 0].grid(True, alpha=0.3)

# 2. Histograma de residuos
axes[0, 1].hist(residuals, bins=20, density=True, alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Distribuci√≥n de Residuos')
axes[0, 1].set_xlabel('Residuos')
axes[0, 1].set_ylabel('Densidad')

# Agregar curva normal
from scipy import stats
x = np.linspace(residuals.min(), residuals.max(), 100)
axes[0, 1].plot(x, stats.norm.pdf(x, residuals.mean(), residuals.std()),
                'r-', label='Normal', linewidth=2)
axes[0, 1].legend()

# 3. Q-Q plot
stats.probplot(residuals, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot')
axes[1, 0].grid(True, alpha=0.3)

# 4. ACF de residuos
plot_acf(residuals, lags=min(20, len(residuals)//4), ax=axes[1, 1], alpha=0.05)
axes[1, 1].set_title('ACF de Residuos')

plt.tight_layout()
plt.show()

# Prueba Ljung-Box
print("\nüìä Prueba Ljung-Box (H0: No hay autocorrelaci√≥n en residuos)")
lb_test = acorr_ljungbox(residuals, lags=min(10, len(residuals)//5), return_df=True)
print(lb_test)

# Prueba de normalidad
print("\nüìä Prueba de Normalidad Jarque-Bera")
jb_stat, jb_pval = stats.jarque_bera(residuals)
print(f"Estad√≠stico JB: {jb_stat:.4f}")
print(f"p-valor: {jb_pval:.4f}")
if jb_pval > 0.05:
    print("‚úÖ No se rechaza normalidad de residuos")
else:
    print("‚ö†Ô∏è Se rechaza normalidad de residuos")

# PASO 8: PRON√ìSTICOS
print("\n" + "=" * 60)
print("PASO 8: PRON√ìSTICOS")
print("=" * 60)

# Solicitar n√∫mero de per√≠odos a pronosticar
print("\n¬øCu√°ntos per√≠odos deseas pronosticar?")
try:
    n_periods = int(input("Ingresa el n√∫mero de per√≠odos (por defecto 8): ") or "8")
except:
    n_periods = 8
    print("Usando valor por defecto: 8 per√≠odos")

# Generar pron√≥sticos
forecast = best_model.forecast(steps=n_periods)

# Crear √≠ndice para pron√≥sticos
if isinstance(df.index, pd.DatetimeIndex):
    forecast_index = pd.date_range(start=df.index[-1] + pd.DateOffset(months=3),
                                   periods=n_periods, freq='Q')
else:
    forecast_index = range(len(df), len(df) + n_periods)

# Transformar de vuelta a escala original
forecast_original = np.exp(forecast)

# Calcular intervalos de confianza
forecast_result = best_model.get_forecast(steps=n_periods)
forecast_ci = forecast_result.conf_int(alpha=0.05)
ci_lower_original = np.exp(forecast_ci.iloc[:, 0])
ci_upper_original = np.exp(forecast_ci.iloc[:, 1])

# Visualizar pron√≥sticos
plt.figure(figsize=(14, 7))

# Determinar cu√°ntos puntos hist√≥ricos mostrar
n_historical = min(40, len(df))

# Datos hist√≥ricos
plt.plot(df.index[-n_historical:], df['PDI'][-n_historical:],
         'o-', label='Datos hist√≥ricos', linewidth=2, markersize=6, color='blue')

# Pron√≥sticos
plt.plot(forecast_index, forecast_original,
         's-', color='red', label='Pron√≥stico', linewidth=2, markersize=8)

# Intervalo de confianza
plt.fill_between(forecast_index, ci_lower_original, ci_upper_original,
                 alpha=0.3, color='red', label='IC 95%')

plt.title(f'Pron√≥stico del IPD - Modelo ARIMA{best_model_spec}', fontsize=16)
plt.xlabel('Per√≠odo', fontsize=12)
plt.ylabel('IPD (Miles de millones de $ 1987)', fontsize=12)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar valores pronosticados
print(f"\nüìä Pron√≥sticos del IPD para los pr√≥ximos {n_periods} per√≠odos:")
print("=" * 70)

forecast_df = pd.DataFrame({
    'Per√≠odo': forecast_index,
    'Pron√≥stico': forecast_original.values,
    'IC_Inferior': ci_lower_original.values,
    'IC_Superior': ci_upper_original.values
})

# Formatear la tabla
forecast_df['Pron√≥stico'] = forecast_df['Pron√≥stico'].round(2)
forecast_df['IC_Inferior'] = forecast_df['IC_Inferior'].round(2)
forecast_df['IC_Superior'] = forecast_df['IC_Superior'].round(2)

print(forecast_df.to_string(index=False))

# Calcular tasas de crecimiento
print("\nüìà Tasas de crecimiento trimestral pronosticadas:")
growth_rates = (forecast_original.values[1:] / forecast_original.values[:-1] - 1) * 100
for i, rate in enumerate(growth_rates):
    print(f"Per√≠odo {i+1} a {i+2}: {rate:.2f}%")

# RESUMEN FINAL
print("\n" + "=" * 60)
print("üéØ RESUMEN DEL AN√ÅLISIS")
print("=" * 60)
print(f"1. Archivo analizado: {filename}")
print(f"2. Serie: Logaritmo del IPD (Ingreso Personal Disponible)")
print(f"3. N√∫mero de observaciones: {len(df)}")
print(f"4. Modelo seleccionado: ARIMA{best_model_spec}")
print(f"5. Criterios de informaci√≥n:")
print(f"   - AIC: {best_model.aic:.2f}")
print(f"   - BIC: {best_model.bic:.2f}")
print(f"6. Pron√≥sticos generados: {n_periods} per√≠odos")
print(f"7. Valor √∫ltimo observado: ${df['PDI'].iloc[-1]:.2f} miles de millones")
print(f"8. Pron√≥stico pr√≥ximo per√≠odo: ${forecast_original.iloc[0]:.2f} miles de millones")

# Opci√≥n para guardar resultados
print("\n" + "=" * 60)
print("üíæ GUARDAR RESULTADOS")
print("=" * 60)
print("\n¬øDeseas descargar los resultados?")
save_option = input("Ingresa 'si' para descargar o presiona Enter para omitir: ").lower()

if save_option == 'si':
    # Preparar archivo de resultados
    results_text = f"""AN√ÅLISIS ARIMA - INGRESO PERSONAL DISPONIBLE (IPD)
{'='*60}

Archivo analizado: {filename}
Fecha del an√°lisis: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}

MODELO SELECCIONADO: ARIMA{best_model_spec}
AIC: {best_model.aic:.2f}
BIC: {best_model.bic:.2f}

PRON√ìSTICOS:
{forecast_df.to_string(index=False)}

RESUMEN DEL MODELO:
{best_model.summary().as_text()}
"""

    # Guardar archivo de texto
    with open('resultados_arima_ipd.txt', 'w') as f:
        f.write(results_text)

    # Guardar pron√≥sticos en CSV
    forecast_df.to_csv('pronosticos_ipd.csv', index=False)

    # Descargar archivos
    files.download('resultados_arima_ipd.txt')
    files.download('pronosticos_ipd.csv')

    print("\n‚úÖ Archivos descargados:")
    print("   - resultados_arima_ipd.txt")
    print("   - pronosticos_ipd.csv")

print("\nüéâ ¬°An√°lisis completado exitosamente!")
```
